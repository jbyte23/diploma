```python
def stacked_bar_senti(df_senti):
  index = ['MMC RTV Slovenija', 'Dnevnik', '24ur.com', 'Siol.net Novice', 'Nova24TV', 'Tednik Demokracija', 'PortalPolitikis']
  positive = []
  neutral = []
  negative = []

  i = 0
  for media in index:
    counts = df_senti.loc[df_senti.media == media, 'sentiment'].value_counts()
    print(counts)
    positive_pct = 0 if 2 not in counts.keys() else counts[2] / counts.sum()
    neutral_pct = 0 if 1 not in counts.keys() else counts[1] / counts.sum()
    negative_pct = 0 if 0 not in counts.keys() else counts[0] / counts.sum()

    positive.append(positive_pct)
    neutral.append(neutral_pct)
    negative.append(negative_pct)

  colors = ['tab:red', 'tab:blue', 'tab:green']

  df_plot = pd.DataFrame({'negativno': negative,
                    'nevtralno': neutral,
                    'pozitivno': positive,
                    }, index=index)
  
  df_plot = df_plot.sort_values('negativno')

  ax = df_plot.plot.bar(stacked=True, color=colors)
```

```python
stacked_bar_senti(df_senti)
```

```python
prob = 0.5   #trump = 0.4
tid = 4
for media in ['Dnevnik', 'MMC RTV Slovenija', '24ur.com', 'Siol.net Novice', 'Nova24TV', 'Tednik Demokracija', 'PortalPolitikis']:
  print(media)
  print()
  for title in df_senti.loc[(df_senti.media == media) & (df_senti.topic_id == tid) & (df_senti.topic_score > prob)].sort_values('topic_score', ascending=False).title[:10]:
    print(title)
  print()
  print('===============================================')

print(len(df_senti.loc[(df_senti.topic_id == tid) & (df_senti.topic_score > prob)]))
```
