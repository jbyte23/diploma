```python
import pandas as pd
import numpy as np
import gensim
from gensim import corpora

import sys
utils_dir = '/content/drive/MyDrive/Diploma'
sys.path.insert(0, utils_dir)
import utils as utils


import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

import pyLDAvis, pyLDAvis.gensim as gensimvis
import matplotlib.pyplot as plt
```

```python
YEAR = 2019

# Po potrebi spremenite pot v naslednji vrstici tako, da pot predstavlja mapo, v kateri so shranjeni model LDA, slovar in korpus
DIR = f'/content/drive/MyDrive/Diploma/lda_models/{YEAR}'

MEDIA_LIST = ['Dnevnik', 'Siol.net Novice', '24ur.com', 'MMC RTV Slovenija', 'Nova24TV', 'Tednik Demokracija', 'PortalPolitikis']
MAINSTREAM_MEDIA_LIST = ['Dnevnik', 'Siol.net Novice', '24ur.com', 'MMC RTV Slovenija']
RIGHT_MEDIA_LIST = ['Nova24TV', 'Tednik Demokracija', 'PortalPolitikis']

# Spremenite naslednjo vrstico, če želite preizkusiti drug model lda.
# Možne vrednosti so: 
# "splosno" - model splošnih tem za izbrano leto
# "slo_politika" - model podtem za temo slovenske politike
# "svet" - model podtem za temo svetovne politike
# "korona" - model podtem za temo koronavirusa (samo 2020)
MODEL_NAME = 'slo_politika'
```

```python
df = pd.read_pickle(f'/content/drive/MyDrive/Diploma/data/2019_{MODEL_NAME}.pkl')

df_mainstream = df.loc[df.media.isin(MAINSTREAM_MEDIA_LIST)]
df_right = df.loc[df.media.isin(RIGHT_MEDIA_LIST)]
df
```

```python
dictionary = corpora.Dictionary.load(f'{DIR}/dictionary_{MODEL_NAME}')
corpus = corpora.MmCorpus(f'{DIR}/corpus_{MODEL_NAME}')
len(dictionary), len(corpus)
```

```python
model_path = f'{DIR}/lda_{MODEL_NAME}'
lda_model =  gensim.models.LdaModel.load(model_path)
```

```python
pyLDAvis.enable_notebook()
topic_data = gensimvis.prepare(lda_model, corpus, dictionary=lda_model.id2word, sort_topics=False)
pyLDAvis.display(topic_data)
```

```python
df_topic_info = utils.get_topic_info(df, topic_data)
df_topic_info = utils.interprete_topics(df_topic_info)
```
